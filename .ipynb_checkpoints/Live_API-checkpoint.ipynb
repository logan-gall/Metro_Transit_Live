This project maps the Metro Transit vehicles in real time. Currently, I have the Blue and Green line light rails displaying in the map.

```python
#imports
import requests
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
from shapely.geometry import Point
import json
import time
from IPython.display import display, clear_output
```

```python
## Initial set up
# API URLs for routes 902 and 901 (Green and Blue Line)
api_url_902 = 'https://svc.metrotransit.org/nextrip/vehicles/902'
api_url_901 = 'https://svc.metrotransit.org/nextrip/vehicles/901'

# Load shapefiles for both the green and blue lines
shapefile_path_green = r'Transit_Shapes/green_line.shp'
shapefile_path_blue = r'Transit_Shapes/blue_line.shp'

# Number of frames for the animation
num_frames = 100
```

```python
# Function to get train data points
def get_train_data(api_url):
    response = requests.get(api_url)
    if response.status_code == 200:
        data = response.json()
        points = [Point(train['longitude'], train['latitude']) for train in data]
        return gpd.GeoDataFrame(data, geometry=points, crs='EPSG:4326')
    else:
        return None
```

```python
# Create the initial plot
fig, ax = plt.subplots(figsize=(10, 10))
gdf_shapefile_green = gpd.read_file(shapefile_path_green)
gdf_shapefile_blue = gpd.read_file(shapefile_path_blue)

# Plot both shapefiles
gdf_shapefile_green.plot(ax=ax, color='green', edgecolor='black')
gdf_shapefile_blue.plot(ax=ax, color='blue', edgecolor='black')

ax.set_title('Metro Transit Live API')
ax.set_axis_off()

# Update the plot in a loop
for _ in range(num_frames):
    # Get data for both routes
    gdf_points_902 = get_train_data(api_url_902)
    gdf_points_901 = get_train_data(api_url_901)

    # Clear the previous plot
    ax.cla()

    # Plot both shapefiles
    gdf_shapefile_green.plot(ax=ax, color='green', edgecolor='black')
    gdf_shapefile_blue.plot(ax=ax, color='blue', edgecolor='black')

    # Plot train data points for both routes
    if gdf_points_902 is not None:
        gdf_points_902.to_crs(gdf_shapefile_green.crs.to_string()).plot(ax=ax, color='green', marker='o')
    if gdf_points_901 is not None:
        gdf_points_901.to_crs(gdf_shapefile_blue.crs.to_string()).plot(ax=ax, color='blue', marker='o')

    # Add a basemap
    ctx.add_basemap(ax, crs=gdf_shapefile_green.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)

    # Customize the plot
    ax.set_title('Metro Transit Live API')
    ax.set_axis_off()

    # Display the updated plot in real-time
    display(fig)
    clear_output(wait=True)

    # Pause for 5 seconds (though I think metro API is every 30-60 seconds)
    time.sleep(5)
```
